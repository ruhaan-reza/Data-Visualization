```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv("https://raw.githubusercontent.com/ruhaan-reza/Data-Visualization/main/IPL.csv")

df.head()
```

how many columns have null values in total

```python
df.isnull().sum()
```

1. Which team won the most matches

```python
df['match_winner'].value_counts()
```

```python
match_wins = df['match_winner'].value_counts()
sns.barplot(y = match_wins.index,x = match_wins.values,palette='viridis')
```

2. Toss Decision Trends

```python
sns.countplot(x=df['toss_decision'])
plt.title('Toss Decision Trends')
```

3. Toss Winner vs Match Winner

```python
count = df[df['toss_winner'] == df['match_winner']]['match_id'].count()
percentage = (count*100)/df.shape[0]
percentage.round(2)
```

4. How do team win?(Runs vs Wickets)

```python
sns.countplot(x=df['won_by'])
```

# Key Player Performances

1 Most 'Player of the Match 'Awards

```python
count = df['player_of_the_match'].value_counts().head(10)
count
```

```python
sns.barplot(y = count.index,x = count.values,palette='mako')
plt.title('Most Player of the Match Awards')
```

Top Scorers

```python
high = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head(10)
high
```

```python
high.plot(kind='bar',color='orange')
plt.title('Top Scorers')
```

**10 Best Bowling Figures**

```python
df['highest_wickets'] = df['best_bowling_figure'].apply(lambda x: x.split('--')[0])
df['highest_wickets'] = df['highest_wickets'].astype(int)
top_bowlers = df.groupby('best_bowling')['highest_wickets'].sum().sort_values(ascending=False).head(10)
top_bowlers
```

```python
top_bowlers.plot(kind='bar',color='green')
plt.title('Top Bowlers')
```

# **Venue Analysis**

**Most Matches Played by Venue**

```python
venue_count = df['venue'].value_counts()
venue_count
```

```python
sns.barplot(y = venue_count.index,x = venue_count.values,palette='mako')
plt.title('Most Matches Played by Venue')
```

**Who won the highest margin by runs?**

```python
df[df['won_by'] == 'Runs'].sort_values(by='margin',ascending=False).head(1)[['match_winner','margin']]
```

**Which player had the highest individual score?**

```python
df[df['highscore'] == df['highscore'].max()][['top_scorer','highscore']]
```

**Which bowler had the best bowling figure?**

```python
df[df['highest_wickets'] == df['highest_wickets'].max()][['best_bowling','best_bowling_figure']]
```
