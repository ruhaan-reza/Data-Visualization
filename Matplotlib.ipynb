```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Matlotlib Basics

```python
x = np.linspace(0,5,11)
x
y = x**2
y
```

```python
plt.plot(x,y)
plt.title("x vs y")
plt.xlabel("x")
plt.ylabel("y")
```

Subplots in Matplotlib

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(y,x,'b')
plt.subplot(2,2,3)
plt.plot(x,y**2,'k')
plt.subplot(2,2,4)
plt.plot(y,x**2,'y')
```

Object Oriented Way

```python
fig = plt.figure(figsize=(20,8),dpi=500)
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y)
axes2 = fig.add_axes([0.3,0.3,0.3,0.3])
axes2.plot(x,y)


```

Types of Plots

```python
plt.scatter(x,y)
```

```python
plt.hist(x)
```

```python
plt.boxplot(x)
```

Saving Plots

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
plt.savefig("my_plot.png")
```
