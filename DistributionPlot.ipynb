```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df
```

```python
df['size'].unique()
```

HistPlot


```python
sns.histplot(df['total_bill'])
```

```python
sns.histplot(df['total_bill'],bins = 10)
```

```python
sns.histplot(df['total_bill'],bins = 10,kde = True)
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins = 10,kde = True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins = 10,kde = True)
```

JointPlot

```python
sns.jointplot(data = df,x = 'total_bill',y = 'tip')
```

```python
sns.jointplot(data = df,x = 'total_bill',y = 'tip', kind='kde')
```

```python
sns.jointplot(data = df,x = 'total_bill',y = 'tip', kind='hex')
```

```python
sns.jointplot(data = df,x = 'total_bill',y = 'tip', kind='reg')
```

PairPlot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue = 'sex')
```

```python
sns.pairplot(df,hue = 'size',palette='rainbow')
```

RugPlot

```python
sns.rugplot(df['total_bill'])
```

```python
sns.rugplot(df)
```
